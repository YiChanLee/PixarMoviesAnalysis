```python
import requests
import bs4
import numpy as np
import pandas as pd
```

```python
url = 'http://www.usinflationcalculator.com/inflation/historical-inflation-rates/'
response = requests.get(url)
html = response.content
print html
```

```python
soup = bs4.BeautifulSoup(html)
```

```python
print(soup.prettify())
```

```python
table = soup.find('tbody')
print table
```

```python
rows = []
for row in table.findAll('tr'):
    rows.append(row)
```

```python
header = [obj.string for obj in rows[0] if obj.string!='\n']
```

```python
data = list()
for row in rows[1:]:
    datum = [obj.string for obj in row if obj.string!='\n']
    data.append(datum)
```

```python
pd.DataFrame(data, columns=header)
```

```python
years = list()
avgs = list()
for datum in data:
    years.append(datum[0])
    avgs.append(datum[-1])
```
